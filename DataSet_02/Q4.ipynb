```python
df_user = df[['user','gender','age_group','job','city','marital']].drop_duplicates().reset_index(drop=True)
df_user['user'].nunique(), len(df_user)
```

```python
df_agg = df.groupby('user')[['prod','purchase']].agg({'prod':'nunique',
                                                     'purchase':'sum'})
df_agg = df_agg.reset_index()
df_agg.head()
```

```python
df_q3 = df_user.merge(df_agg, on = 'user', how = 'inner')
df_q3 = df_q3.drop(columns = 'user')
```

```python
df_q3 = df_q3['gender'].replace({'M':1,'F':0})
```

```python
ser_repl = pd.Series(range(7), index = arr_u)
```

```python
pd.crosstab(df_q3['age_group'], df_q3['age_group'].replace(ser_repl))
```

```python
df_3 = df.fillna(0)
df_3['prod_cat1'] = df_3['prod_cat1'].astype(int)
df_3['prod_cat2']=df_3['prod_cat2'].astype(int)
df_3['prod_cat3']=df_3['prod_cat3'].astype(int)
df_3['prod'] = df_3['prod_cat1'].astype(str) + '-' + df_3['prod_cat2'].astype(str) +'-' + df_3['prod_cat3'].astype(str)
df_3.head()
```

```python
df_33 = df_3.set_index(['user','prod'])
#df_3.loc[:,'prod_count'] = df_3.groupby('user').value_counts(['prod'])
```

```python
df_33.loc[:,'prod_count'] = df_3.groupby('user').value_counts(['prod'])
```

```python
df_33 = df_33.reset_index()
```

```python
df_33.head()
```

```python
df_a = df_33[['gender','age_group','job','city','marital','prod_count','purchase']]
df_a.head()
```

```python
df_a['age_group'].unique()
```

```python
df_a['age_group'] = df_a['age_group'].apply(lambda x : 0 if x == '0-17'else 
                       (1 if x == '18-25' else 
                       (2 if x == '26-35' else 
                       (3 if x == '36-45' else 
                       (4 if x == '46-50' else 
                       (5 if x == '51-55' else 6)))))
                       )
```

```python
df_a['gender'] = df_a['gender'].apply(lambda x : 0 if x == 'F' else 1)
```

```python
df_a = pd.get_dummies(df_a,columns = ['job','city'])
```

```python
df_a
```

```python
len(df_a.columns)
```

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_a)
```

```python
kmeans = KMeans(n_clusters=7, random_state = 123)
```

```python
kmeans.fit(df_scaled)
```

```python
labels = kmeans.labels_
```

```python
silhouette_score(df_scaled, labels)
```

```python
df_scaled['cluster'] = labels
df_scaled['cluster'].value_counts() # 각 군집의 크기
```
