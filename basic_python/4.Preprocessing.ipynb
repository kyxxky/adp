```python
def udf_sum(x,y):
    result = x + y
    return result
```

```python
udf_sum(pd.Series([2,4,6]),4)
```

```python
def udf_sum2(x,y):
    result = x.sum() + y
    return result
```

```python
udf_sum2(pd.Series([2,4,6]),4)
```

```python
def udf_mean(x,digit=3):
    result = round(x.mean(),digit)
    return result
```

```python
udf_mean(pd.Series([4,6,8,17,0.0072321]), digit = 5)
udf_mean(pd.Series([4,6,8,17,0.0072321]))
```

```python
#입력객체의 원소를 모두 제곱한 값의 합에 제곱근을 취하는 사용자 정의함수를 만들었을 때, 입력 [3,5,9,20] 의 출력값은
def udf_euclidean_dist(x):
    result = np.sqrt(sum(x.pow(2))) #x.pow(2).sum()**0.5
    return result
udf_euclidean_dist(pd.Series([3,5,9,20]))
```

```python
#Standardization udf를 만들고 [-4,5,7,9] 를 입력한 경우, 7은 어떤 값으로 반환되는가
def udf_normalization(x):
    result = (x-x.mean())/x.std()
    return result
udf_normalization(pd.Series([-4,5,7,9]))
```

```python
#minmax 정규화 udf를 만들고 [-4,5,7,9]를 입력한 경우, 7은 어떤 값으로 반환되는가
def udf_minmax(x):
    result = (x - x.min())/(x.max()-x.min())
    return result
udf_minmax(pd.Series([-4,5,7,9]))
```
